```python
import subprocess
import concurrent.futures

commands = [
    'python mrmd3.0.py -i ./encoded_cksaap_0.4_with_label.csv -o merged_encode_cksaap_0.4_out.csv -g true',
    'python mrmd3.0.py -i ./encoded_cksaap_0.5_with_label.csv -o merged_encode_cksaap_0.5_out.csv -g true',
    'python mrmd3.0.py -i ./encoded_cksaap_0.6_with_label.csv -o merged_encode_cksaap_0.6_out.csv -g true',
    'python mrmd3.0.py -i ./encoded_cksaap_0.7_with_label.csv -o merged_encode_cksaap_0.7_out.csv -g true'
]

def run_command(command):
    try:
        subprocess.run(command, shell=True, check=True)
        print(f"Command {command} completed successfully.")
    except subprocess.CalledProcessError as e:
        print(f"Error running command {command}: {e}")

with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(run_command, commands)
```
